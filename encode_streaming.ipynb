```python
def rev(a):
    
    if a == None or len(a) ==0:
        return False
    
    x = a.split(' ')
    output = []
    
    for i in x[::-1]:
        output.append(i)
    
    return ' '.join(output)

a = "I hope I get this job"

print(rev(a))
```

```python
def word(a):
    return ' '.join(a.split(' ')[::-1])

a = "I hope I get this job"

print(word(a))
```
